```python
# Loading Data:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Running Locally, yakinkan "file_" berada di folder "data"
dfJ = pd.read_csv("jakarta.csv", low_memory = False, encoding='utf8')
dfD = pd.read_csv("depok.csv", low_memory = False, encoding='utf8')
dfT = pd.read_csv("tangerang.csv", low_memory = False, encoding='utf8')

dfJ.shape, dfD.shape, dfT.shape
```

```python
#Menghilangkan data yang bernilai NaN
dfJ.dropna(axis=1, how='all', inplace=True)
dfJ
```

```python
#Mereset Index
dfJ.reset_index(inplace=True)
dfJ
```

```python
dfJ.dropna(axis=0, thresh=4, inplace=True)
dfJ
```

```python
#Melihat persebaran missing value pada data kota Jakarta
sns.heatmap(dfJ.isnull(),yticklabels=False,cbar=False)
```

# Interpretasi
Pada output kota Jakarta diatas, terdapat missing value di beberapa variabel, terlihat di variabel garasi dan hadap mempunyai missing value terbanyak, sedangkan untuk variabel carport, sertifikat dan deskripsi juga memiliki missing value tetapi tidak sebanyak variabel garasi dan hadap


```python
# Mengatasi missing value pada variabel garasi, carport dan listrik menggunakan nilai modus.
modus = dfJ['garasi'].mode()[0]
dfJ['garasi'] = dfJ['garasi'].fillna(modus)
modus1 = dfJ['carport'].mode()[0]
dfJ['carport'] = dfJ['carport'].fillna(modus1)
modus2 = dfJ['listrik'].mode()[0]
dfJ['listrik'] = dfJ['listrik'].fillna(modus2)
```

```python
#Melihat persebaran missing value setelah beberapa missing value pada variabel diatasi.
sns.heatmap(dfJ.isnull(),yticklabels=False,cbar=False)
```

```python
#Mengatasi missing value pada variabel hadap menggunakan nilai modus dan sertifikat menggunakan (nilai forward fill).
moduss = dfJ['hadap'].mode()[0]
dfJ['hadap'] = dfJ['hadap'].fillna(moduss)
dfJ.fillna(method='ffill', inplace=True)
```

```python
#Melihat persebaran missing value setelah diatasi.
sns.heatmap(dfJ.isnull(),yticklabels=False,cbar=False)
```

```python
dfJ.info()
```

```python
#Mengubah tipe data.
dfJ['LT'] = dfJ['LT'].astype('float64')
dfJ['KT'] = dfJ['KT'].astype('Int64')
dfJ['KM'] = dfJ['KM'].astype('Int64')
dfJ['garasi'] = dfJ['garasi'].astype('Int64')
dfJ['carport'] = dfJ['carport'].astype('Int64')
dfJ['lokasi'] = dfJ['lokasi'].astype('category')
dfJ['sertifikat'] = dfJ['sertifikat'].astype('category')
dfJ['hadap'] = dfJ['hadap'].astype('category')
dfJ['listrik'] = dfJ['listrik'].astype('Int64')
```
